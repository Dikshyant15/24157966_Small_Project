```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import scrfft
```

```python
def scrfft(xdata,ydata):
    sdata=np.argsort(xdata)
    xdatas=xdata[sdata]
    ydatas=ydata[sdata]
    
    
    xmin=np.min(xdata)
    xmax=np.max(xdata)
    ndata=len(xdata)
    x=(xmax-xmin)/(ndata-1)*np.arange(ndata)+xmin
    y=np.interp(x,xdatas,ydatas)
    
    yf = 2.0*np.fft.rfft(y)/(ndata+1)
    a=np.real(yf)
    b=-np.imag(yf)
    yf=0
    
    a[0]=0.5*a[0]
    
    f=np.arange(len(a))/(xmax-xmin)
    
    return(f, a, b) #frequency, amplitude sin and amplitude cosine values 
```

```python
newdata = np.genfromtxt('./star.csv')
mydate = newdata[:,0]
myint = newdata[:,1]
```

```python
plt.figure(0,dpi=100)
plt.plot(mydate,myint,marker='o')
plt.xlabel('Time, Days')
plt.ylabel('Intensity')
```

```python
freqs, ccoef, scoef = scrfft.scrfft(mydate,myint)
```

```python
pwr = ccoef**2+scoef**2
```

```python
plt.figure(1,dpi=100)
plt.plot(1/freqs,pwr)
plt.xlabel('Time period, days')
plt.ylabel('Power')
```
